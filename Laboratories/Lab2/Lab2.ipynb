# Task 0
Declare a simple class that represents a `Vehicle`. The class should have an `__init__` function which sets
* name
* max_speed
* mileage

````python
class Vehicle:
    color='White'
    def __init__(self, name='Audi', max_speed=250,mileage=0):
        self.name=name
        self.max_speed=max_speed
        self.mileage=mileage
    
    def seating_capacity(self,cap=5):
        self.capacity=cap
    
    def fare_cost(self):
        return self.capacity*100
    
Samochodzik=Vehicle()
print(Samochodzik.max_speed)
````

## Task 5
Implement a class `case`. The class should store a creation date and four values $E1$, $E2$, $E3$ and $E4$. Write a function (not a method) to randomly generate dates and another one to generate values $E1-E4$. The second function should return the values as tuple. Choose ranges for the values.

Use:
```
import random

```
Create a number of instances of `case`. Use `filter` and `sort` for to select values fulfilling a certain condition form and sort them according to creation date.

````python
import random
from datetime import date, timedelta

def losuj_date(data_poczatkowa, data_koncowa):
    # 1. Obliczamy rozpiętość czasu (tzw. delta) między datami
    roznica = data_koncowa - data_poczatkowa
    # Roznica.days zwraca nam całkowitą liczbę dni w tym przedziale
    ile_dni = roznica.days
    # 2. Losujemy liczbę dni od 0 do maksymalnej liczby dni w przedziale
    losowe_dni = random.randrange(ile_dni)
    # 3. Dodajemy wylosowaną liczbę dni do daty początkowej
    wylosowana_data = data_poczatkowa + timedelta(days=losowe_dni)
    
    return wylosowana_data

start = date(1970, 1, 1)
koniec = date(2026, 3, 6)


def gen_values(min,max):
    lista=[]
    for i in range(0,4):
        lista.append(random.randint(min,max))

    krotka = tuple(lista)
    return krotka

E=gen_values(0,69)
wynik = losuj_date(start, koniec)
print(f"Wylosowana data kalendarzowa: {wynik},E={E}")

class case:
    def __init__(self):
        self.creation=losuj_date(start,koniec)
        self.E=gen_values(69,420)

    def __repr__(self):
        # This makes printing the object readable instead of showing a memory address
        return f"Case(Date: {self.creation}, E: {self.E})"

# A=case()
# B=case()
# c=case()
# D=case()
# E=case()
# F=case()

lista_obiektow = [case() for _ in range(15)]

filtered_objects = list(filter(lambda x: sum(x.E) > 1000, lista_obiektow))
print("Obiekty z sumą E większą niż 1000:", filtered_objects)
filt_and_sort=sorted(filtered_objects, key=lambda x: x.creation)
print("Obiekty:", filt_and_sort)
````
