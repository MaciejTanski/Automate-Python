Procedural and object oriented paradigms of programming.  
Procedural programing centers around developing procedures (functions) and passing data in between while performing programmed tasks. So in a sense data and functionality remain separated.

Object Oriented Programing (OOP) in Python is based on a couple of concepts, we will now discuss:
* Class and object
* Methods and attributes
* Encapsulation
* Inheritance and polymorphism

Note: OOP, allows to do some things in a more natural and understandable manner, but there should be limits to the excessive use. What I mean, it is good when it is good, and sometimes to much is just to much.

## Encapsulation or hermetization
One of the crucial concepts in OPP is encapsulation or hermetization. That is the ability to limit access to some of the object's attributes and making them unavailable for modification or even reading. Various languages offer different levels at which such restriction is possible (e.g. **C++** has a sophisticated suite of possibilities). In Python this is limited to a naming convention. Any attribute that starts with \_\_ (two low bars?) is restricted. Let's see an example of a Cat's bowl, that we are going to make unavailable:

````python
class Cat:
    def __init__(self):
        self.__bowl = 'empty' # cats have bowls with food
    def FillBowl(self):
        self.__bowl = "Filled with food"
    def Eat(self):
        self.__bowl = 'empty'
    def Print(self):
        print(self.__bowl)
        cathappy = "Cat is happy" if self.__bowl != 'empty' else "Cat is sad"
        print(cathappy)
````

````python
a = Cat()
a.FillBowl()
````

````python
print(a.__bowl)
````

````python
a.Print()
````

````python
a.__bowl = 'empty'
````

````python
print(a.__bowl)
````

````python
a.Print()
````

````python
a.Eat()
a.Print()
````

We note that we can not access Cat's bowl for reading (AttributeError),
and if we try to assign to it it has no effect on the attribute.

## Inheritance and polymorphism

Inheritance is central to OOP. The concept is based on extending functionality of existing classes by deriving from them new ones, that conceptually perform similar, but more specific tasks. So inheritance introduces hierarchy to classes form abstract concepts (such as a general shape in geometry) down to very specific ones (for shapes, a square or a triangle, etc.).

The class that is more abstract, from which we derive is called the **base** class (general shape), and the less abstract class formed via inheritance is the **derived** one (a square is less general than a shape, so it could be derived from it).

Concept of inheritance allows for:
* Less code, since code is reused
* Possibly more order in the code
* Polymorphism, i.e. handling objects in a unified way

Let us present the concept with

```sequence
Mammal<-Dog
      <-Cat
```
Note: Show step-by-step, not all at once:

````python
class Mammal():
    def __init__(self):
        self.drinks_milk = 'Yes'
````

````python
class Cat(Mammal):
    pass
class Dog(Mammal):
    pass
````

````python
m = Mammal()
c = Cat()
d = Dog()
````

Finally we get something like this:

````python
class Mammal():
    def __init__(self, n):
        self.name = n
        self.drinks_milk = "Yes"
    def make_noise(self): # all derived classes make noise 
        print(self.noise) # similar to abstract in C++
    def print_name(self):
        print('Has no name')
````

````python
m = Mammal('generic mamal')
m.print_name()
m.make_noise()
````

````python
class Cat(Mammal):
    def __init__(self):
        self.noise = "Meow"
        Mammal.__init__(self, "Cat")
    def print_name(self):
        print('Behemoth')
````

````python
class Dog(Mammal):
    def __init__(self):
        self.noise = "Bark"
        Mammal.__init__(self, "Cat")
    def print_name(self):
        print('Cerberus')
````

````python
m = Mammal('')
print(m.drinks_milk)
m.print_name()
# m.make_noise()
````

````python
c = Cat()
print(c.drinks_milk)
c.print_name()
c.make_noise()
````

# isinstance vs type
Sometimes there is a need to verify a type of an object. In OOP inheritance is based on the derived type extending the base one. This can be understood that the derived type object is the base and some more.  
Sometimes one might need to check if a class or an object of a class are of specific type, or inherit from one.

* `type(obj)` returns the type of an object. Can be used with `is` to check if object is of specific type. Does not work on inheritance.
* `isinstance(obj, type)` returns true if an object is derived from a class passed as an argument. Works on inheritance.
* `issubclass(type1, type2)` Works on types, not objects. Checks if a class has been derived from the base class.

Let's see in an example

````python
class Animal:
    phylum = 'not defined'
    __isextinct = 'No' # ??

    def __init__(self, phy):
        self.phylum = phy

    def no_letters(self):
        return len(self.phylum)-1

    def SetExtinct(self, val):
        self.__isextinct = val
    def GetExtinct(self):
        return self.__isextinct

    def IsWarm(self):
        return self.isWarm
````

````python
class Mamal(Animal):
    isWarm = 'Yes'
    def __init__(self):
        Animal.__init__(self, 'ssaki')

class Reptile(Animal):
    isWarm = 'No'
    def __init__(self):
        Animal.__init__(self, 'gady')
````

````python
class Dog(Mamal):
    def __init__(self):
        Mamal.__init__(self)

class Cat(Mamal):
    def __init__(self):
        Mamal.__init__(self)

class Snake(Reptile):
    def __init__(self):
        Reptile.__init__(self)

class Lizard(Reptile):
    def __init__(self):
        Reptile.__init__(self)
````

````python
dog = Dog()
cat = Cat()
snake = Snake()
viper = Snake()
lizard = Lizard()
````

`type(obj)`

````python
print(type(dog))
print(type(snake))
````

````python
print(type(dog) is Dog)
print(type(snake) is Dog)
````

````python
print(type(dog) is Mamal)
print(type(snake) is Reptile)
print(type(snake) is Animal)
````

`isinstance(obj, type)`

````python
print(isinstance(cat, Animal))
print(isinstance(cat, Mamal))
print(isinstance(cat, Reptile))
print(isinstance(cat, Cat))
````

````python
print(isinstance(snake, Animal))
print(isinstance(snake, Mamal))
print(isinstance(snake, Reptile))
print(isinstance(snake, Cat))
print(isinstance(snake, Lizard))
````

````python
print(isinstance(viper, type(snake)))
````

````python
print(issubclass(Cat, Animal))
print(issubclass(Cat, Mamal))
print(issubclass(Cat, Reptile))
print(issubclass(Cat, (Reptile, Dog, Mamal)))
````

````python
print(type(dog) is Dog)
print(type(cat) is Mamal)
print(type(snake) is Animal)
print(type(lizard) is Lizard)
````
